# FIFA19データ解析

まず、'data.xlsx'からデータを取り込んで上から１０個目までのデータを表示します。

```python
import pandas as pd
input_book = pd.ExcelFile('data.xlsx')
input_sheet_df = input_book.parse(input_book.sheet_names[0])
input_sheet_df.head(10)
```

null値がいくつあるかを確認します。

```python
input_sheet_df.isnull().sum()
```

つぎに、各列の基本統計量を表示してデータの特徴を確認します。<br>
Wage２はmeanがmediumを大きく上回っており、給料の上位層の選手が平均を釣り上げているだろうことが考えられます。

```python
input_sheet_df.describe()
```

分析のため、null値を０で埋めます。

```python
input_sheet_df.fillna(0).head(10)
```

ドリブルの巧さと給料の関係に注目しました。

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import scipy.stats
Dribbling = input_sheet_df.loc[:,'Dribbling']
Wage2 = input_sheet_df.loc[:,'Wage2']
result = scipy.stats.linregress(Dribbling,Wage2)
result_slope = result.slope
result_intercept = result.intercept
plt.plot(Dribbling, [result_slope * u + result_intercept for u in Dribbling])
plt.scatter(Dribbling,Wage2)
plt.title('Wage and Dribbling of Players in FIFA19')
plt.ylabel('Wage')
plt.xlabel('Dribbling')
plt.show()
```

双峰型に近い分布をしています。<br>
dribbling　の値が０〜２０で給料の高いデータはおそらくGKのデータだと考えられるので、GKのデータを取り除いたデータフレームを抽出します。

```python
feild_dframe=input_sheet_df[input_sheet_df.Position != 'GK']
```

また、dribbling　の値が０のものを取り除いたデータフレームを作ります。さらに他のnull値は0で埋めます。

```python
feild_dframe_2=feild_dframe[feild_dframe.Dribbling != 0].fillna(0)
```

```python
import matplotlib.pyplot as plt
import scipy.stats
wage2 = feild_dframe_2.loc[:,'Wage2']
dribbling = feild_dframe_2.loc[:,'Dribbling']
finishing = feild_dframe_2.loc[:,'Finishing']

result = scipy.stats.linregress(dribbling,wage2)
print('傾き=', result.slope.round(4), '切片=', result.intercept.round(4), \
     '信頼係数=', result.rvalue.round(4), 'p値=', result.pvalue.round(4), \
     '標準誤差=', result.stderr.round(4))
result_slope = result.slope
result_intercept = result.intercept
plt.plot(dribbling, [result_slope * u + result_intercept for u in dribbling])
plt.scatter(dribbling,wage2)
plt.title('Wage and Dribbling of Players in FIFA19')
plt.ylabel('Wage')
plt.xlabel('Dribbling')
plt.show()
```

傾きは 0.5694で、視覚的にもWageとDribblingに緩やかな正の相関が見られます。<br>
ただ、一部の高給とりが傾きを引き上げている感は否めません。

jointplot関数でも散布図を描いてみます。

```python
import seaborn as sns
sns.jointplot(dribbling,wage2)
```

先ほどより上位層がまばらに分布しているのが強調されてます。<br>
Wage2が１５０以下の部分を見ると、Dribblingが60以上であればあまりドリブルのうまさは賃金に大きな影響を及ばさなそうです。

wage２ を被説明変数にして、dribbling を説明変数にした単回帰の結果を示します。 p値を観る限り有意ですが, 修正R二乗値を観る限り説明力は低いです。

```python
import statsmodels.api as sm
model = sm.OLS(wage2, sm.add_constant(dribbling))
results = model.fit()
print(results.summary())
print('p-values\n', results.pvalues)
```

同様に、wage を被説明変数にして、finishing を説明変数にした単回帰の結果を示します。 これもp値を観る限り有意ですが、 修正R二乗値を観る限り説明力はかなり低そうです。

```python
import statsmodels.api as sm
model = sm.OLS(wage2, sm.add_constant(finishing))
results = model.fit()
print(results.summary())
print('p-values\n', results.pvalues)
```

MessiやCR7などWageの高すぎる上位層の影響を取り除いてみたいので、<br>
まず、percentilesで指定してより細かく賃金の分位点を見てみます。

```python
feild_dframe_2['Wage2'].describe(percentiles=[.1,.2,.3,.4,.5,.6,.7,.8,.9])
```

ここで、Wage2が２５以下である選手を対象にこれまで同様の分析をします。すると、傾きが0.1476まで小さくなりました。

```python
feild_dframe_wage90=feild_dframe_2[feild_dframe_2['Wage2']<=25]
import matplotlib.pyplot as plt
import scipy.stats
wage2 = feild_dframe_wage90.loc[:,'Wage2']
dribbling = feild_dframe_wage90.loc[:,'Dribbling']
finishing = feild_dframe_wage90.loc[:,'Finishing']

result = scipy.stats.linregress(dribbling,wage2)
print('傾き=', result.slope.round(4), '切片=', result.intercept.round(4), \
     '信頼係数=', result.rvalue.round(4), 'p値=', result.pvalue.round(4), \
     '標準誤差=', result.stderr.round(4))
result_slope = result.slope
result_intercept = result.intercept
plt.plot(dribbling, [result_slope * u + result_intercept for u in dribbling])
plt.scatter(dribbling,wage2)
plt.title('Wage and Dribbling of Players in FIFA19')
plt.ylabel('Wage')
plt.xlabel('Dribbling')
plt.show()
```

次は攻撃的なポジションの選手（CMより前線）に限定して分析してみます。

```python
result = scipy.stats.linregress(finishing,wage2)
print('傾き=', result.slope.round(4), '切片=', result.intercept.round(4), \
     '信頼係数=', result.rvalue.round(4), 'p値=', result.pvalue.round(4), \
     '標準誤差=', result.stderr.round(4))
result_slope = result.slope
result_intercept = result.intercept
plt.plot(finishing, [result_slope * u + result_intercept for u in finishing])
plt.scatter(finishing,wage2)
plt.title('Wage and Dribbling of Attacking Players in FIFA19')
plt.ylabel('Wage')
plt.xlabel('Drbbling')
plt.show()
```

相変わらず上位層の影響は強そうですが、Dribbling、Finishing共に傾きが大きくなりました。

フィールドプレイヤー全体の時より、攻撃的ポジションの選手に限定するとR２乗値も共に大きくなりました。

最後に、Wage2を被説明変数、Dribbling、Finishing、Vision、Ballcontrol、FKAccuracyを説明変数として重回帰分析を実行します。

```python
import numpy as np
from sklearn import linear_model, datasets
from sklearn.linear_model import LinearRegression

wage = fw_dframe.loc[:,'Wage2']
dribbling = fw_dframe.loc[:,'Dribbling']
finishing = fw_dframe.loc[:,'Finishing']
vision = fw_dframe.loc[:,'Vision']
ballcontrol = fw_dframe.loc[:,'BallControl']
fkaccuracy = fw_dframe.loc[:,'FKAccuracy']

equation_df=pd.concat([wage,dribbling,finishing,vision,ballcontrol,fkaccuracy], axis=1)
wage2 = pd.DataFrame(equation_df.Wage2)
x_list = equation_df.drop("Wage2",1)

#異常値やnull になっている値を除去する
x_list = x_list.drop(x_list.columns[np.isnan(x_list).any()], axis=1)

model = sm.OLS(wage2, sm.add_constant(x_list))
result =model.fit()
print(result.summary())
print(result.pvalues)
```

ドリブルの巧さとフィニッシュの精度、視野の広さ、ボールコントロールはプラスに有意に働くことが分かりました。FKの精度はややマイナスの係数ですが、有意であるとは言えなさそうです。<br>
単回帰の時よりR2乗値は大きくなりました。

```python
from statsmodels.stats.outliers_influence import *
num_cols = model.exog.shape[1] 
print(num_cols)
vifs = [variance_inflation_factor(model.exog, i) for i in range(0, num_cols)]
pdv = pd.DataFrame(vifs, index=model.exog_names, columns=["VIF"])
print(pdv)
```

VIF値をみると、どの変数も１０を超えていないので、多重共線性は発生していないと考えられます。
